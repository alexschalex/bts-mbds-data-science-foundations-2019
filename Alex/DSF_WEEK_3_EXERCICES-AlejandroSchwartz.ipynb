# EXERCISE 1: image filtering [OBLIGATORY]

From the datasets (images) available from [skimage data](https://scikit-image.org/docs/dev/api/skimage.data.html), choose an **RGB** image and do the following:

* Crop the image
* Smoothen the image with a kernel filter.
* Sharpen the image with a kernel filter
* Find the edges of the image with at least two different methods.
* From the processes above, save the image that you prefer as a .jpg



```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
%matplotlib inline
```

```python
from skimage.data import retina

retina=retina()

plt.imshow(retina)
plt.show()
```

* Crop the image


```python
plt.imshow(retina[400:800,0:750])
plt.show()
```

```python
retina.shape
```

* Smoothen the image with a kernel filter.


```python
from scipy import ndimage



kernel = np.array([[[1/9,1/9,1/9]], 
                    [[1/9,1/9,1/9]], 
                  [[1/9,1/9,1/9]]]) 

filtered = ndimage.convolve(retina, kernel, mode='constant', cval=0.0)

fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(10, 10))

ax[0].imshow(retina, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(filtered, cmap=plt.cm.gray)
ax[1].set_title('Smoothen Pic')

plt.tight_layout()
plt.show()
```

* Sharpen the image with a kernel filter


```python

kernel = np.array([ [[0, -1, 0]], 
                    [[-1, 3, -1]], 
                    [[0, -1, 0]]])


filtered = ndimage.convolve(retina, kernel, mode='mirror', cval=0.0)

fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(10, 10))

ax[0].imshow(retina, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(filtered, cmap=plt.cm.gray)
ax[1].set_title('Sharpen convolution')

plt.tight_layout()
plt.show()
```

* Find the edges of the image with at least two different methods.


```python




kernel_edge = np.array([[ [1, 0, -1]], 
                       [[0, 0, 0]], 
                        [[-1, 0, 1]]]) #Edge Detection Filter


edge = ndimage.convolve(retina, kernel_edge, mode='mirror', cval=0.0)

fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(10, 10))

ax[0].imshow(retina, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(edge, cmap=plt.cm.gray)
ax[1].set_title('Edge convolution')

plt.tight_layout()
plt.show()
```

```python
kernel = np.array([[ [-1, -1, -1]], 
                    [[-1, 8, -1]], 
                    [[-1, -1, -1]]]) / 9


filtered = ndimage.convolve(retina, kernel, mode='constant', cval=0.0)

fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(10, 10))

ax[0].imshow(retina, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(filtered, cmap=plt.cm.gray)
ax[1].set_title('High pass filter convolution')

plt.tight_layout()
plt.show()
```

* From the processes above, save the image that you prefer as a .jpg


```python
imageio.imwrite('retina.jpg',retina)
```

# Working With B&W image

```python
from skimage.data import retina
from skimage.color import rgb2gray


retina=retina()

grayret=rgb2gray(retina)


fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(retina)
ax[0].set_title("Original")
ax[1].imshow(grayret, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

```python
grayret.shape
```

* Crop the image


```python
plt.imshow(grayret[400:800,0:750],cmap=plt.cm.gray)
plt.show()
```

* Smoothen the image with a kernel filter.


```python
from scipy import ndimage



kernel = np.array([[1/9,1/9,1/9], 
                    [1/9,1/9,1/9], 
                  [1/9,1/9,1/9]]) 

filtered = ndimage.convolve(grayret, kernel, mode='constant', cval=0.0)

fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(10, 10))

ax[0].imshow(grayret, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(filtered, cmap=plt.cm.gray)
ax[1].set_title('Smoothen Pic')

plt.tight_layout()
plt.show()
```

* Sharpen the image with a kernel filter


```python

kernel = np.array([ [0, -1, 0], 
                    [-1, 3, -1], 
                    [0, -1, 0]])


filtered = ndimage.convolve(grayret, kernel, mode='mirror', cval=0.0)

fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(10, 10))

ax[0].imshow(grayret, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(filtered, cmap=plt.cm.gray)
ax[1].set_title('Sharpen convolution')

plt.tight_layout()
plt.show()
```

* Find the edges of the image with at least two different methods.


```python

kernel_edge = np.array([[1, 1, 1], 
                       [0, 0, 0], 
                        [-1, -1, -1]]) #Horizontal Edge Detection Filter


edge = ndimage.convolve(grayret, kernel_edge, mode='mirror', cval=0.0)

fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(10, 10))

ax[0].imshow(grayret, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(edge, cmap=plt.cm.gray)
ax[1].set_title('Horizontal Edge convolution')

plt.tight_layout()
plt.show()
```

```python
kernel = np.array([ [-1, -1, -1], 
                    [-1, 8, -1], 
                    [-1, -1, -1]]) / 9


filtered = ndimage.convolve(grayret, kernel, mode='constant', cval=0.0)

fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(10, 10))

ax[0].imshow(grayret, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(filtered, cmap=plt.cm.gray)
ax[1].set_title('High pass filter convolution')

plt.tight_layout()
plt.show()
```

* From the processes above, save the image that you prefer as a .jpg


```python
imageio.imsave('grayret.jpg',grayret)
```

* Try to fix the image using opening/closing algorithms


```python
plt.imshow(graycos)
plt.colorbar()
plt.show()
```

```python
markers=np.zeros_like(graycos)

markers[graycos>=0.45]=1
markers[graycos<0.45]=0
plt.imshow(markers)
plt.colorbar()
```

```python
fill_cos=ndimage.binary_fill_holes(markers)
plt.imshow(fill_cos)
plt.show()
```

```python
obj,label=ndimage.label(fill_cos)
print(label)
plt.imshow(obj)
plt.colorbar()
```

```python
open_cos=ndimage.binary_opening(fill_cos,structure=np.ones((9,9)))

plt.imshow(open_cos,cmap=plt.cm.gray)
plt.show()
```

```python
closed_cos=ndimage.binary_closing(open_cos, structure=np.ones((9,9)))

plt.imshow(closed_cos,cmap=plt.cm.gray)
plt.show()
```

```python
fill_cos2=ndimage.binary_fill_holes(closed_cos)
plt.imshow(fill_cos2)
plt.show()
```

```python
obj,label=ndimage.label(fill_cos2)
print(label)
plt.imshow(obj)
plt.colorbar()
```

```python
open_cos=ndimage.binary_opening(fill_cos2,structure=np.ones((30,30)))

plt.imshow(open_cos,cmap=plt.cm.gray)
plt.show()
```

```python
obj,label=ndimage.label(open_cos)
print(label)
plt.imshow(obj)
plt.colorbar()
```
